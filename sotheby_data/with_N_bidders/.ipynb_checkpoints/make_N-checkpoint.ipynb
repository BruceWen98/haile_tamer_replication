```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sotheby_stats = pd.read_csv("../sothebys_stats.csv")
bidder_numbers = pd.read_csv("/Users/brucewen/Desktop/honors_thesis/youtube_data/bidder_assignment/bidders_ALL35sotheby_v2.csv")

bidder_numbers['global_ID'] = bidder_numbers['auction_ID']+bidder_numbers['lot'].apply(str)
sotheby_stats['global_ID'] = sotheby_stats['auction_ID']+sotheby_stats['lot'].apply(str)
```

```python
bidder_numbers = bidder_numbers[['global_ID','num_uniq_bidders']]
bidder_numbers
```

```python
def num_bidders(total_bids,num_uniq_bidders):
    true_num_bidders = num_uniq_bidders
    if num_uniq_bidders>total_bids:
        true_num_bidders = total_bids
    if num_uniq_bidders == 0 or num_uniq_bidders == 1:
        true_num_bidders = min([2,total_bids])
    return true_num_bidders
```

```python
ids_withNumBidders = np.unique(bidder_numbers["global_ID"])
sotheby_stats = sotheby_stats[sotheby_stats["global_ID"].isin(ids_withNumBidders)]

sotheby_stats = pd.merge(bidder_numbers, sotheby_stats, on='global_ID', how='right')
sotheby_stats['N_bidders'] = sotheby_stats.apply(lambda x: num_bidders(x.total_bids, x.num_uniq_bidders), axis=1)
```

```python
def is_outlier(points, thresh=3.5):
    """
    Returns a boolean array with True if points are outliers and False 
    otherwise.

    Parameters:
    -----------
        points : An numobservations by numdimensions array of observations
        thresh : The modified z-score to use as a threshold. Observations with
            a modified z-score (based on the median absolute deviation) greater
            than this value will be classified as outliers.

    Returns:
    --------
        mask : A numobservations-length boolean array.

    References:
    ----------
        Boris Iglewicz and David Hoaglin (1993), "Volume 16: How to Detect and
        Handle Outliers", The ASQC Basic References in Quality Control:
        Statistical Techniques, Edward F. Mykytka, Ph.D., Editor. 
    """
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh

x = sotheby_stats["seg1_final_price"]
filtered = x[~is_outlier(x)]

plt.hist(filtered)
```

```python
bins3 = [80000,650000]
sotheby_stats_lowVal = sotheby_stats[sotheby_stats['seg1_final_price']<=bins3[0]]
sotheby_stats_midVal = sotheby_stats[(sotheby_stats['seg1_final_price']<=bins3[1]) & (sotheby_stats['seg1_final_price']>bins3[0])]
sotheby_stats_highVal = sotheby_stats[sotheby_stats['seg1_final_price']>bins3[1]]
```

```python
sotheby_stats_lowVal.to_csv("sotheby_stats_lowVal.csv")
sotheby_stats_midVal.to_csv("sotheby_stats_midVal.csv")
sotheby_stats_highVal.to_csv("sotheby_stats_highVal.csv")
```
